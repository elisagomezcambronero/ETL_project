```python
#idsubgenre
```

```python
def id_sql(Id_column, Id):
    df_id=pd.DataFrame(Id_column.unique())
    id = range(1, len(df_id) + 1)
    df_id[f"Id{Id}"]= id
    df_id.rename(columns = {0: Id}, inplace = True)
    
    return df_id
```

```python
df_subgenre=id_sql(df["Subgenre"], "Subgenre")
```

```python
df=df.merge(df_subgenre, on= "Subgenre", how="left")
```

```python
df["IdGenre_Subgenre"]=df["Idgenre"]+df["IdSubgenre"]
```

```python
#idlocation
df_loc=df.groupby(by=['Event_Country', 'Event_City']).size().reset_index()
```

```python
df_loc.drop(0, axis=1, inplace=True)
```

```python
df_genre.rename(columns = {0: "Genre"}, inplace = True)
```

```python
id = range(1, len(df_loc) + 1)
df_loc["Idloc"]= id
```

```python
df=df.merge(df_loc, on= "Event_City", how="left")
```

```python
df.drop("Event_Country_y", axis=1, inplace=True)
```

```python
df.rename(columns = {"Event_Country_x": "Event_Country"}, inplace = True)
```

```python
#id long lat
df_coor=id_sql(df["Event_Longitude"], "Event_Longitude")
```

```python
df=df.merge(df_coor, on= "Event_Longitude", how="left")
```

# TABLA SONGS

```python
df["Song_Name"] = df["Song_Name"].str.replace('"', "")
```

```python
for index, row in df.iterrows():
    song_id = engine.execute(f"""select idSongs from songs WHERE idSongs = "{row['Idsongs']}";""")
    
    if len(list(song_id)) > 0: 
        print(f"lo siento, este id ya existe")
    else: 
        engine.execute(f"""
            INSERT INTO songs (idSongs, Song_Name, Artist, Release_Date, Song_Decade) VALUES
            ("{row["Idsongs"]}","{row['Song_Name']}","{row["Artist"]}", "{row['Release_Date']}", "{row['Song_Decade']}");""")
```

```python
for index, row in df.iterrows():
    song_id = engine.execute(f"""select idSong_Genre from song_genre WHERE idSong_Genre = "{row['Idgenre']}";""")
    
    if len(list(song_id)) > 0: 
        "lo siento, este id ya existe"
    else: 
        engine.execute(f"""
            INSERT INTO song_genre (idSong_Genre, Genre) VALUES
            ("{row["Idgenre"]}","{row['Genre']}");""")
```

```python
for index, row in df.iterrows():
    song_id = engine.execute(f"""select idsong_subgenres from song_subgenres WHERE idsong_subgenres = "{row['IdSubgenre']}";""")
    
    if len(list(song_id)) > 0: 
        "lo siento, este id ya existe"
    else: 
        engine.execute(f"""
            INSERT INTO song_subgenres (idsong_subgenres, Subgenre, Subgenre2, song_genre_idSong_Genre) VALUES
            ("{row["IdSubgenre"]}","{row['Subgenre']}", "{row['Subgenre_2']}", "{row['Idgenre']}");""")
        
```

```python
for index, row in df.iterrows():
    song_id = engine.execute(f"""select idEventLocation from event_location WHERE idEventLocation = "{row['Idloc']}";""")
    
    if len(list(song_id)) > 0: 
        "lo siento, este id ya existe"
    else: 
        engine.execute(f"""
            INSERT INTO event_location (idEventLocation, Event_Country, Event_City) VALUES
            ("{row["Idloc"]}","{row['Event_Country']}", "{row['Event_City']}");""")
```

```python
for index, row in df.iterrows():
    song_id = engine.execute(f"""select idEvent_Coordenates from event_coordenates WHERE idEvent_Coordenates = "{row['IdEvent_Longitude']}";""")
    
    if len(list(song_id)) > 0: 
        "lo siento, este id ya existe"
    else: 
        engine.execute(f"""
            INSERT INTO event_coordenates (idEvent_Coordenates, Event_Latitude, Event_Longitude, event_location_idEventLocation) VALUES
            ("{row["IdEvent_Longitude"]}","{row['Event_Longitude']}", "{row['Event_Latitude']}", "{row['Idloc']}");""")
```
